```python
drivewheels = dataframe['drive-wheels'].value_counts().to_frame()
```

```python
drivewheels.reset_index(inplace = True)
drivewheels = drivewheels.rename(columns = {"drive-wheels":"value_counts"})
```

```python
drivewheels.index.name = "drive-wheels"
drivewheels.head()
```

```python
enginelocation = dataframe['engine-location'].value_counts().to_frame()
enginelocation.reset_index(inplace = True)
enginelocation = enginelocation.rename(columns = {"engine-location":"value_counts"})
enginelocation.index.name = 'engine-location'
enginelocation.head()
```

```python
dataframe['drive-wheels'].unique()
```

```python
groupframe = dataframe[['drive-wheels','body-style','price']]
groupframe.head()
```

```python
dw_gframe = groupframe.groupby(['drive-wheels'] , as_index = False).agg({'price':'mean'})
dw_gframe.head()
```

```python
dw_bs_gframe = groupframe.groupby(['drive-wheels','body-style'] , as_index = False).mean()
dw_bs_gframe.head()
```

```python
dw_bs_pframe = dw_bs_gframe.pivot(index = "drive-wheels" , columns = "body-style").fillna(0)
dw_bs_pframe.reset_index()
```

```python
dw_bs_pframe.head()
```

```python
bs_gframe = groupframe.groupby(['body-style'], as_index=False).agg({'price':'mean'})
bs_gframe.head()
```

```python
import matplotlib.pyplot as plot
```

```python
plot.pcolor(dw_bs_pframe, cmap = 'RdBu')
plot.xlabel("Body-Style")
plot.ylabel("Drive-Wheels")
plot.colorbar()
plot.show()
```

```python
dw_bs_pframe.head()
```

```python
fig , ax = plot.subplots()
hmap = ax.pcolor(dw_bs_pframe , cmap = 'RdBu')
row_labels = dw_bs_pframe.columns.levels[1]
col_labels = dw_bs_pframe.index

ax.set_xticks(np.arange(dw_bs_pframe.shape[1])+0.5 , minor = False)
ax.set_yticks(np.arange(dw_bs_pframe.shape[0])+0.5 , minor = False)
ax.set_xticklabels(row_labels , minor = False)
ax.set_yticklabels(col_labels , minor = False)

plot.xticks(rotation = 90)
fig.colorbar(hmap)
plot.show()
```

```python
from scipy import stats
```

```python
def printpearson(x):
    pearson_cof , pval = stats.pearsonr(dataframe[x] , dataframe['price'])
    print(pearson_cof , pval)
```
