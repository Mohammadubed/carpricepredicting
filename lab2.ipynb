```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
dataframe = pd.read_csv(url)
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
dataframe.columns=headers
dataframe.head()
```

```python
dataframe.replace("?" , np.nan , inplace = True)
```

```python
dataframe.head()
```

```python
dataframe2 = dataframe.isnull()
```

```python
dataframe2.head(5)
```

```python
dataframe3 = dataframe2.notnull()
```

```python
for column in dataframe2.columns.values.tolist():
    print(column)
    print(dataframe2[column].value_counts())
    print("")
```

```python
for column in ["normalized-losses","stroke","bore","horsepower","peak-rpm"]:
    mean = dataframe[column].astype("float").mean(axis=0)
    dataframe[column] =dataframe[column].replace(np.nan, mean)
```

```python
dataframe.head()
```

```python
common = dataframe["num-of-doors"].value_counts().idxmax()
dataframe["num-of-doors"] = dataframe["num-of-doors"].replace(np.nan,common)
```

```python
dataframe["num-of-doors"].head()
```

```python
dataframe.dtypes
```

```python
dataframe[["bore","stroke","price","peak-rpm"]] = dataframe[["bore","stroke","price","peak-rpm"]].astype("float")
```

```python
dataframe[["normalized-losses"]] = dataframe[["normalized-losses"]].astype("float")
```

```python
dataframe.dtypes
```

```python
dataframe[["city-mpg","highway-mpg"]] = 235/dataframe[["city-mpg","highway-mpg"]] 
```

```python
dataframe.rename(columns = {"city-mpg":"city-L/100km","highway-mpg":"highway-L/100km"} , inplace=True)
```

```python
dataframe.head()
```

```python
for column in ["length","height","width"]:
    dataframe[column] = dataframe[column]/dataframe[column].max()
```

```python
dataframe.head()
```

```python
dataframe["horsepower"] = dataframe["horsepower"].astype(int , copy=True)
```

```python
import matplotlib as plot
```

```python
plot.pyplot.hist(dataframe["horsepower"])
plot.pyplot.xlabel("horsepower")
plot.pyplot.ylabel("count")
plot.pyplot.title("horsepower bins")
```

```python
bins = np.linspace(min(dataframe["horsepower"]), max(dataframe["horsepower"]),4)
group_name = ["Low" , "Medium" , "High"]
dataframe["horsepower-binned"] = pd.cut(dataframe["horsepower"] , bins , labels = group_name , include_lowest = True)
dataframe[["horsepower","horsepower-binned"]].head(20)
```

```python
dataframe["horsepower-binned"].value_counts()
```

```python
plot.pyplot.hist(dataframe["horsepower-binned"])
plot.pyplot.xlabel("horsepower")
plot.pyplot.ylabel("counts")
plot.pyplot.title("horsepower-binned")
```
